PDF and CDF of Exponential Distribution 

```python
import matplotlib.pyplot as plt
import numpy as np

def exp_pdf(x,lam):
    return lam * np.exp(-lam * x)

def exp_cdf(x,lam):
    return 1 - np.exp(-lam * x)

def mean(lam):
    return 1/lam

def median(lam):
    return np.log(2)/lam

def mode(lamb):
    return 0

def variance(lam):
    return 1/(lam)**2



x = np.linspace(0,5,1000)

fig, ax = plt.subplots()
ax.plot(x,exp_pdf(x,1))
ax.set_title("Exponential pdf (lambda = 1) ")
plt.show()
print("mean is ",mean(1))
print("median is ",median(1))
print("mode is ",mode(1))
print("variance is ",variance(1))

fig, ax = plt.subplots()
ax.plot(x,exp_pdf(x,2))
ax.set_title("Exponential pdf (lambda = 2) ")
plt.show()
print("mean is ",mean(2))
print("median is ",median(2))
print("mode is ",mode(2))
print("variance is ",variance(2))

fig, ax = plt.subplots()
ax.plot(x,exp_pdf(x,3))
ax.set_title("Exponential pdf (lambda = 3) ")
plt.show()
print("mean is ",mean(3))
print("median is ",median(3))
print("mode is ",mode(3))
print("variance is ",variance(3))

fig, ax = plt.subplots()
ax.plot(x,exp_cdf(x,1))
ax.set_title("Exponential cdf (lambda = 1) ")

fig, ax = plt.subplots()
ax.plot(x,exp_cdf(x,2))
ax.set_title("Exponential cdf (lambda = 2) ")

fig, ax = plt.subplots()
ax.plot(x,exp_cdf(x,3))
ax.set_title("Exponential cdf (lambda = 3) ")
plt.show()
```

d) It's a Poisson Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
k = np.linspace(1,20,20,dtype=int)
print(k)

def factorial(x):
     if x < 0:
        return 0
     elif x == 0 or x == 1:
        return 1
     else:
        fact = 1
        while(x > 1):
            fact *= x
            x -= 1
        return fact
             
def poisson_pmf(lamb,k):
    return ((lamb**k) * (np.exp(-lamb))) / factorial(k)

def mean(lamb):
    return lamb

def variance(lamb):
    return lamb

fig , ax = plt.subplots()
ax.set_title("PMF of Poisson Distribution")
ax.set_ylabel("Probabilty")
ax.set_xlabel("value of k")
ax.plot(1,poisson_pmf(5,1),"*b")
ax.plot(2,poisson_pmf(5,2),"*b")
ax.plot(3,poisson_pmf(5,3),"*b")
ax.plot(4,poisson_pmf(5,4),"*b")
ax.plot(5,poisson_pmf(5,5),"*b")
ax.plot(6,poisson_pmf(5,6),"*b")
ax.plot(7,poisson_pmf(5,7),"*b")
ax.plot(8,poisson_pmf(5,8),"*b")
ax.plot(9,poisson_pmf(5,9),"*b")
ax.plot(10,poisson_pmf(5,10),"*b")
ax.plot(11,poisson_pmf(5,11),"*b")
ax.plot(12,poisson_pmf(5,12),"*b")
plt.show()

print("Mean is : ",mean(5))
print("Variance is : ",variance(5))
```
